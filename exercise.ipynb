##### Valid Sudoku

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
s
```

```python
s.sum()
```

```python
b = np.sum(s, axis = 1) # sum of all column elements
for i in b:
    if i != 45:
        print("suduko is not valid")
else :
        print("suduko is valid")
```

```python
c = np.sum(s, axis = 0) # sum of all row elements
for i in c:
    if i != 45:
        print("suduko is not valid")
else :
    print("suduko is valid")
```

```python
for i in range(0, 9, 3):
    for j in range(0, 9, 3):
        n = s[i:i+3, j:j+3]
        n.sum
        print(n.sum())
```

### General Questions

```python
import numpy as np
```

```python
# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

##### Get the shape of the matrix.

```python
data.shape
```

##### Find the average age of students.

```python
slice = data[:, 0]
slice
```

```python
slice.mean()
```

### OR

```python
avg_age = data[:, 0].mean()
print(avg_age)
```

##### Extract Math marks of all students.

```python
data
```

```python
maths = data[:, 1]
maths
```

### OR

```python
maths_marks = data[:, 1]
print(maths_marks)
```

##### Find the highest Science mark.

```python
science = data[0 : 5, 2]
science
```

```python
science.max()
```

### OR

```python
highest_science = data[:, 2].max()
print(highest_science)
```

##### Get details of the student who scored more than 90 in Math.

```python
data
```

```python
student = data[data[:, 1] > 90]
print(student)
```

##### Increase Math marks of all students by 5.

```python
maths_5 = data[:, 1] + 5
print(maths_5)
```

##### Find how many students are younger than 19.

```python
age = data[:, 0]
age
```

```python
count = 0
for i in age:
    if i < 19:
        count += 1
print(count)
```

### OR

```python
age_less_than_19 = (data[:, 0] < 19).sum()
print(age_less_than_19)
```

##### Calculate the average marks in each subject (column-wise mean).

```python
marks_maths = data[:, 1]
marks_science = data[:, 2]
marks_maths
marks_science
```

```python
print(marks_maths.mean())
print(marks_science.mean())
```

### OR

```python
avg_marks = data[:, 1:3].mean(axis=0)
print(avg_marks)
```

##### Get data of students who scored at least 80 in both subjects.

```python
data
```

```python
students_both = data[(data[:, 1] >= 80) & (data[:, 2] >= 80)]
print(students_both)
```

##### Replace all Science marks < 75 with 0.

```python
data[data[:, 2] < 75, 2] = 0
print(data)
```
