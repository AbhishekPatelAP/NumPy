```python
arr[0] # row slicing through indexing
```

```python
arr[4]
```

```python
arr[0, 0]
```

```python
arr[0, 4]
```

```python
arr[5, 0]
```

```python
arr[5, 4]
```

```python
slice_2378 = arr[0:2, 1:3]
slice_2378
```

```python
slice_192024252930 = arr[3:6, 3:5]
slice_192024252930
```

```python
slice_192024252930 = arr[3:, 3:]
slice_192024252930
```

Boolean Indexing

```python
arr = np.arange(11, 21)
arr
```

```python
bool_index = arr % 2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
