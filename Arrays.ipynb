Array Methods

```python
arr.min() # will give minimum value from the matrix
```

```python
arr.max() # will give maximum value from the matrix
```

```python
arr.sum()
```

```python
sum = int(arr.sum())
print(sum)
```

```python
np.sum(arr, axis= 0) # column sum
```

```python
np.sum(arr, axis = 1) # row sum
```

```python
arr.mean() # average
```

```python
arr.std() # standered deviation
```

```python
arr.argmax() # max element at which index
```

```python
arr.argmin() # min element at which index
```
