```python
a2 - a1
```

```python
a1 * a2
```

```python
a1 / a2
```

```python
a1 // a2
```

#### Broadcasting

```python
l  = [10, 20, 30, 40, 50]
arr = np.array(l)
```

```python
arr + 10 # will add 10 in all elements of list l
```

```python
arr * 1.2 # as bonus
```

```python
arr2 = np.arange(1, 26).reshape(5, 5)
arr2
```

```python
arr2 + 10 # we can add 10 to the matrix as well
```

```python
arr2 * 10 # can multiply by 10 to all elements of array or matrics
```

```python
A
```

```python
B
```

```python
A @ B
```

### Advance Array Manipulations :

#### Stacking Array

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
```

```python
A
```

```python
B
```

```python
# we can not stack two metrics
```

```python
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
```

```python
np.vstack((a, b)) # vertical stack
```

```python
np.hstack((a, b)) # horizontal stack
```

```python
np.column_stack((a, b)) # column stack
```
